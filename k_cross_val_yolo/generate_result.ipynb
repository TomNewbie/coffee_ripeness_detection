```python
import pandas as pd
import json
```

```python
filename = 'results_yolov8.json'
with open(filename, 'r') as file:
    data = json.load(file)
metric_df = pd.DataFrame.from_dict(data)
visualize_metric = ['mean', 'std', 'min', 'max']
metric_df.describe().loc[visualize_metric]
```
